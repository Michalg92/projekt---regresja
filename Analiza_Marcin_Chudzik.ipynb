```python
from sklearn import linear_model

x, y = np.array(df['grade']).reshape(-1, 1), df['price']
```

```python
x = x[:, np.newaxis, 0]

x_train = x[:-20]
x_test = x[-20:]

y_train = y[:-20]
y_test = y[-20:]

model = linear_model.LinearRegression()

model.fit(x_train, y_train)

y_pred_train1 = model.predict(x_train)
y_pred_test1 = model.predict(x_test)

score1 = r2_score(y_test, y_pred_test1)
print(f"Score 1 = {score1}")

bias_train=mean_absolute_percentage_error(y_train, y_pred_train1)
print(f"Bias train = {bias_train}")

bias_test=mean_absolute_percentage_error(y_test, y_pred_test1)
print(f"Bias test = {bias_test}")
```

```python
print("GRADE -- PRICE")

x = np.array(df['grade']).reshape(-1, 1)
y = df['price']

x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=123)

model_lin1=LinearRegression()
model_lin1.fit(x_train, y_train)

y_pred_train1=model_lin1.predict(x_train)
y_pred_test1=model_lin1.predict(x_test)

score1 = r2_score(y_test, y_pred_test1)

print(f"Score 1 = {score1}\n")

bias_train=mean_absolute_percentage_error(y_train, y_pred_train1)
print(f"Bias train = {bias_train}")

bias_test=mean_absolute_percentage_error(y_test, y_pred_test1)
print(f"Bias test = {bias_test}")

model_tree=DecisionTreeRegressor()
model_tree.fit(x_train, y_train)

y_pred_test2=model_tree.predict(x_test)

r2_score(y_test, y_pred_test2)

d= model_tree.get_depth()
l = model_tree.get_n_leaves()
print(f"\nDecission Tree \nDepth = {d}, Leaves = {l}\n")

model_tree=DecisionTreeRegressor()
model_tree.fit(x_train, y_train)

score2 = r2_score(y_test, y_pred_test2)
print(f"Score 2 = {score2}")
```
